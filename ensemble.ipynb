```python
from PIL import Image

#run this if images are not saved locally as np array

#in order for this to work, you need to download the data from kaggle as a zip and upload it to the following directory:
# /content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/data/

glioma_path = '/content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/data/Brain_Cancer raw MRI data/Brain_Cancer/brain_glioma'
menin_path = '/content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/data/Brain_Cancer raw MRI data/Brain_Cancer/brain_menin'
tumor_path = '/content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/data/Brain_Cancer raw MRI data/Brain_Cancer/brain_tumor'

image_data = []
image_labels = []

target_size = (256, 256) # this is a hyper param, rescaled size of images that will be used as the input shape for the CNN



for directory_path in [glioma_path, menin_path, tumor_path]:
    if os.path.exists(directory_path):
        label = os.path.basename(directory_path) # directory corresponds to label
        for filename in os.listdir(directory_path):
            full_path = os.path.join(directory_path, filename)
            if os.path.isfile(full_path):
                try:
                    img = Image.open(full_path).convert('RGB')
                    img_resized = img.resize(target_size) # <- we resize so the image data isn't incredibly huge but still provides enough information to train on
                    img_array = np.asarray(img_resized)
                    image_data.append(img_array)
                    image_labels.append(label)
                except Exception as e:
                    print(f"Error processing {full_path}: {e}")
    else:
        print(f"Directory not found: {directory_path}")

# Convert the lists to numpy arrays
image_data = np.array(image_data)
image_labels = np.array(image_labels)

print("Image data shape:", image_data.shape)
print("Image labels shape:", image_labels.shape)
print("Example labels:", image_labels[:10])
```

```python
# defining models

ACTIVATION = 'gelu'
# DENSE_DIM = 32
# INPUT_LAYER_DIM = 64
# CONVOLUTION_LAYER_DIM = 32
# INPUT_LAYER = layers.Conv2D(INPUT_LAYER_DIM, (3, 3), activation=ACTIVATION, input_shape=(256, 256, 3))
# CONVOLUTION_LAYER = layers.Conv2D(CONVOLUTION_LAYER_DIM, (3, 3), activation = ACTIVATION)
# POOLING_LAYER = layers.MaxPooling2D((2, 2))
# DENSE_LAYER = layers.dense(DENSE_DIM, ACTIVATION)
# OUTPUT_LAYER = layers.dense(len(class_names))

model_1 = models.Sequential()
model_1.add(layers.Conv2D(64, (3, 3), activation=ACTIVATION, input_shape=(256, 256, 3))) # input is of the form height, width, num colors
model_1.add(layers.MaxPooling2D((2, 2)))
model_1.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_1.add(layers.MaxPooling2D((2, 2)))
model_1.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_1.add(layers.MaxPooling2D((2, 2)))
model_1.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_1.add(layers.Flatten())
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(len(class_names)))

model_2 = models.Sequential()
model_2.add(layers.Conv2D(64, (3, 3), activation=ACTIVATION, input_shape=(256, 256, 3)))
model_2.add(layers.MaxPooling2D((2, 2)))
model_2.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_2.add(layers.MaxPooling2D((2, 2)))
model_2.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_2.add(layers.MaxPooling2D((2, 2)))
model_2.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_2.add(layers.Flatten())
model_2.add(layers.Dense(32, activation=ACTIVATION))
model_2.add(layers.Dense(32, activation=ACTIVATION))
model_2.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_2.add(layers.Dense(len(class_names)))

model_3 = models.Sequential()
model_3.add(layers.Conv2D(64, (3, 3), activation=ACTIVATION, input_shape=(256, 256, 3)))
model_3.add(layers.MaxPooling2D((2, 2)))
model_3.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_3.add(layers.MaxPooling2D((2, 2)))
model_3.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_3.add(layers.MaxPooling2D((2, 2)))
model_3.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_3.add(layers.Flatten())
model_3.add(layers.Dense(32, activation=ACTIVATION))
model_3.add(layers.Dense(32, activation=ACTIVATION))
model_3.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_3.add(layers.Dense(len(class_names)))

model_4 = models.Sequential()
model_4.add(layers.Conv2D(64, (3, 3), activation=ACTIVATION, input_shape=(256, 256, 3)))
model_4.add(layers.MaxPooling2D((2, 2)))
model_4.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_4.add(layers.MaxPooling2D((2, 2)))
model_4.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_4.add(layers.MaxPooling2D((2, 2)))
model_4.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_4.add(layers.Flatten())
model_4.add(layers.Dense(32, activation=ACTIVATION))
model_4.add(layers.Dense(32, activation=ACTIVATION))
model_4.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_4.add(layers.Dense(len(class_names)))

model_5 = models.Sequential()
model_5.add(layers.Conv2D(64, (3, 3), activation=ACTIVATION, input_shape=(256, 256, 3)))
model_5.add(layers.MaxPooling2D((2, 2)))
model_5.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_5.add(layers.MaxPooling2D((2, 2)))
model_5.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_5.add(layers.MaxPooling2D((2, 2)))
model_5.add(layers.Conv2D(32, (3, 3), activation=ACTIVATION))
model_5.add(layers.Flatten())
model_5.add(layers.Dense(32, activation=ACTIVATION))
model_5.add(layers.Dense(32, activation=ACTIVATION))
model_5.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_1.add(layers.Dense(32, activation=ACTIVATION))
model_5.add(layers.Dense(len(class_names)))
```

```python
#run this cell to change image labels to ints (CNN wants this and won't work without it)

# change labels from strings to int

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the string labels to integer labels
y_train_encoded_1 = label_encoder.fit_transform(y_train_1)
y_test_encoded_1 = label_encoder.transform(y_test_1)
y_train_encoded_2 = label_encoder.fit_transform(y_train_2)
y_test_encoded_2 = label_encoder.fit_transform(y_test_2)
y_train_encoded_3 = label_encoder.fit_transform(y_train_3)
y_test_encoded_3 = label_encoder.fit_transform(y_test_3)
y_train_encoded_4 = label_encoder.fit_transform(y_train_4)
y_test_encoded_4 = label_encoder.fit_transform(y_test_4)
y_train_encoded_5 = label_encoder.fit_transform(y_train_5)
y_test_encoded_5 = label_encoder.fit_transform(y_test_5)



# Update y_train and y_test to use the encoded labels
y_train_1 = y_train_encoded_1
y_test_1 = y_test_encoded_1
y_train_2 = y_train_encoded_2
y_test_2 = y_test_encoded_2
y_train_3 = y_train_encoded_3
y_test_3 = y_test_encoded_3
y_train_4 = y_train_encoded_4
y_test_4 = y_test_encoded_4
y_train_5 = y_train_encoded_5
y_test_5 = y_test_encoded_5
```

```python
# training model

# only run this if weights are not saved locally
# this will take an extraordinarily long time (> 1 hour if you don't have colab pro)
# if you do have colab pro, T4 gpus will speed this up quite a bit, should take ~1 minute.

# Disable eager execution
tf.config.run_functions_eagerly(False)

EPOCHS = 30
OPTIMIZER = tf.keras.optimizers.SGD(
    learning_rate=0.02,
    momentum=0.9,
    nesterov=False,
    weight_decay=None,
    clipnorm=None,
    clipvalue=None,
    global_clipnorm=None,
    use_ema=False,
    ema_momentum=0.99,
    ema_overwrite_frequency=None,
    loss_scale_factor=None,
    gradient_accumulation_steps=None,
    name='SGD'
)

model_1.compile(optimizer=OPTIMIZER,
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model_2.compile(optimizer=OPTIMIZER,
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model_3.compile(optimizer=OPTIMIZER,
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model_4.compile(optimizer=OPTIMIZER,
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model_5.compile(optimizer=OPTIMIZER,
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Convert NumPy arrays to TensorFlow tensors
x_train_1_tensor = tf.convert_to_tensor(x_train_1, dtype=tf.float32)
y_train_1_tensor = tf.convert_to_tensor(y_train_1, dtype=tf.int32)
x_test_1_tensor = tf.convert_to_tensor(x_test_1, dtype=tf.float32)
y_test_1_tensor = tf.convert_to_tensor(y_test_1, dtype=tf.int32)

x_train_2_tensor = tf.convert_to_tensor(x_train_2, dtype=tf.float32)
y_train_2_tensor = tf.convert_to_tensor(y_train_2, dtype=tf.int32)
x_test_2_tensor = tf.convert_to_tensor(x_test_2, dtype=tf.float32)
y_test_2_tensor = tf.convert_to_tensor(y_test_2, dtype=tf.int32)

x_train_3_tensor = tf.convert_to_tensor(x_train_3, dtype=tf.float32)
y_train_3_tensor = tf.convert_to_tensor(y_train_3, dtype=tf.int32)
x_test_3_tensor = tf.convert_to_tensor(x_test_3, dtype=tf.float32)
y_test_3_tensor = tf.convert_to_tensor(y_test_3, dtype=tf.int32)

x_train_4_tensor = tf.convert_to_tensor(x_train_4, dtype=tf.float32)
y_train_4_tensor = tf.convert_to_tensor(y_train_4, dtype=tf.int32)
x_test_4_tensor = tf.convert_to_tensor(x_test_4, dtype=tf.float32)
y_test_4_tensor = tf.convert_to_tensor(y_test_4, dtype=tf.int32)

x_train_5_tensor = tf.convert_to_tensor(x_train_5, dtype=tf.float32)
y_train_5_tensor = tf.convert_to_tensor(y_train_5, dtype=tf.int32)
x_test_5_tensor = tf.convert_to_tensor(x_test_5, dtype=tf.float32)
y_test_5_tensor = tf.convert_to_tensor(y_test_5, dtype=tf.int32)


history_1 = model_1.fit(x_train_1_tensor, y_train_1_tensor, epochs=EPOCHS,
                    validation_data=(x_test_1_tensor, y_test_1_tensor))

history_2 = model_2.fit(x_train_2_tensor, y_train_2_tensor, epochs=EPOCHS,
                    validation_data=(x_test_2_tensor, y_test_2_tensor))

history_3 = model_3.fit(x_train_3_tensor, y_train_3_tensor, epochs=EPOCHS,
                    validation_data=(x_test_3_tensor, y_test_3_tensor))

history_4 = model_4.fit(x_train_4_tensor, y_train_4_tensor, epochs=EPOCHS,
                    validation_data=(x_test_4_tensor, y_test_4_tensor))

history_5 = model_5.fit(x_train_5_tensor, y_train_5_tensor, epochs=EPOCHS,
                    validation_data=(x_test_5_tensor, y_test_5_tensor))

# metrics + evaluation

plt.plot(history_1.history['accuracy'], label='m1 train')
plt.plot(history_1.history['val_accuracy'], label = 'm1 test')

plt.plot(history_2.history['accuracy'], label='m2 train')
plt.plot(history_2.history['val_accuracy'], label = 'm2 test')

plt.plot(history_3.history['accuracy'], label='m3 train')
plt.plot(history_3.history['val_accuracy'], label = 'm3 test')

plt.plot(history_4.history['accuracy'], label='m4 train')
plt.plot(history_4.history['val_accuracy'], label = 'm4 test')

plt.plot(history_5.history['accuracy'], label='m5 train')
plt.plot(history_5.history['val_accuracy'], label = 'm5 test')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.0, 1])
plt.legend(loc='lower right')

checkpoint_path_1 = "/content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/checkpoints_1.weights.h5"
model_1.save_weights(checkpoint_path_1)

checkpoint_path_2 = "/content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/checkpoints_2.weights.h5"
model_2.save_weights(checkpoint_path_2)

checkpoint_path_3 = "/content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/checkpoints_3.weights.h5"
model_3.save_weights(checkpoint_path_3)

checkpoint_path_4 = "/content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/checkpoints_4.weights.h5"
model_4.save_weights(checkpoint_path_4)

checkpoint_path_5 = "/content/drive/MyDrive/CS3540/CS3540-Assignments/final-project/checkpoints_5.weights.h5"
model_5.save_weights(checkpoint_path_5)

test_loss, test_acc = model_1.evaluate(x_test_1_tensor,  y_test_1_tensor, verbose=2)

print("accuracy on test dataset: ", test_acc)

test_loss, test_acc = model_2.evaluate(x_test_2_tensor,  y_test_2_tensor, verbose=2)

print("accuracy on test dataset: ", test_acc)

test_loss, test_acc = model_3.evaluate(x_test_3_tensor,  y_test_3_tensor, verbose=2)

print("accuracy on test dataset: ", test_acc)

test_loss, test_acc = model_4.evaluate(x_test_4_tensor,  y_test_4_tensor, verbose=2)

print("accuracy on test dataset: ", test_acc)

test_loss, test_acc = model_5.evaluate(x_test_5_tensor,  y_test_5_tensor, verbose=2)

print("accuracy on test dataset: ", test_acc)
```

```python
from collections import Counter
from scipy import stats

y_true = np.concatenate((y_test_1, y_test_2, y_test_3, y_test_4, y_test_5))
x = np.concatenate((x_test_1, x_test_2, x_test_3, x_test_4, x_test_5))

# Predict on the entire test set with each model
predictions_1 = np.argmax(model_1.predict(x), axis=1)
predictions_2 = np.argmax(model_2.predict(x), axis=1)
predictions_3 = np.argmax(model_3.predict(x), axis=1)
predictions_4 = np.argmax(model_4.predict(x), axis=1)
predictions_5 = np.argmax(model_5.predict(x), axis=1)

# Stack the predictions to easily find the mode for each image
all_predictions = np.stack([predictions_1, predictions_2, predictions_3, predictions_4, predictions_5], axis=1)

# Calculate the mode for each image across all models
y_pred, _ = stats.mode(all_predictions, axis=1)

# Flatten y_pred as stats.mode can return a 2D array
y_pred = y_pred.flatten()

accuracy = 0
for i in range(len(y_true)):
  if y_true[i] == y_pred[i]:
    accuracy += 1

print("Ensemble accuracy:", accuracy / len(y_true))
```
